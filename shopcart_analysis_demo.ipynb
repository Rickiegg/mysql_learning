```python
query = 'SELECT * FROM bread_basket'
# data = pd.read_sql_query(query, engine)
data = pd.read_sql(query, engine)
data
```

```python
# data=pd.read_csv('/Users/rickie/Desktop/mysql_learning/bread_basket.csv',header=None)
# data.columns=['Date','Time','Transaction','Item']
data
# 统一小写
data['Item'] = data['Item'].str.lower()
data
```

```python
# 去掉 none 项
data = data.drop(data[data.Item == 'none'].index)
```

```python
# 得到一维数组 orders_series，并且将 Transaction 作为 index, value 为 Item 取值
orders_series = data.set_index('Transaction')['Item']
orders_series
```

```python
# 将数据集进行格式转换
transactions = []
temp_index = 0
for i, v in orders_series.items():
    if i != temp_index:
       temp_set = set()# set去重
       temp_index = i
       temp_set.add(v)
       transactions.append(temp_set)
    else:
       temp_set.add(v)
transactions
```

```python
# 挖掘频繁项集和频繁规则
itemsets, rules = apriori(transactions, min_support=0.02,  min_confidence=0.5)
print('频繁项集：', itemsets)
print('关联规则：', rules)
```
