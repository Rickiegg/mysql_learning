```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from collections import Counter
```

```python
order=pd.read_csv('/Users/rickie/Desktop/czbank_test/olist_orders_dataset.csv')
cust=pd.read_csv('/Users/rickie/Desktop/czbank_test/olist_customers_dataset.csv')
prd=pd.read_csv('/Users/rickie/Desktop/czbank_test/olist_order_items_dataset.csv')

prd
```

```python
data=pd.merge(order,cust,on='customer_id')
data=pd.merge(data,prd,on='order_id')
data.head()
```

```python
data.dropna(inplace=True)
data.info()
```

```python
price_m=data['price'].mean()
data.customer_city.value_counts()
```

```python
data.groupby('customer_id').price.mean().reset_index().price
```

```python
def if_high(x):
    if x>=price_m:
        return 'high'
    else:
        return 'average'
data['type']=data.groupby('customer_id').price.mean().reset_index().price.apply(if_high)
```

```python

cnt=data.pivot_table('order_id',columns='type',index=['customer_state'],aggfunc='count')
cnt['total']=cnt['high']+cnt['average']
cnt['high_pct']=cnt['high']/cnt['total']

# cnt['high_pct'] = cnt['high_pct'].apply(lambda x: format(x, '.2%')) 
cnt
cnt['total_pct']=cnt['high'].sum()/cnt['total'].sum()
cnt.info()
```

```python
cnt['tgi']=cnt['high_pct']/cnt['total_pct']*100
tgi=cnt[cnt['total']>cnt['total'].median()]
tgi.sort_values(by='tgi',ascending=False)
```
