```python
import pandas as pd
import seaborn as sns
from datetime import datetime,date
import matplotlib.pyplot as plt
#import pyecharts.options as opts
#from pyecharts.charts import Bar, Line
%matplotlib inline
import numpy as np
```

```python
data=pd.read_excel('/Users/rickie/Desktop/mysql_learning/RFM_data.xlsx')
data.head()
```

```python
data=data[['买家昵称','付款日期']]
data.columns=['cust_id','pay_time']
```

```python
# data['pay_date']=data['pay_time'].map(lambda x:x.strftime('%Y-%m-%d'))
```

```python
# data['首单日期']
min_date=data.groupby(['cust_id'])['pay_time'].min().reset_index()
min_date.rename(columns={'pay_time':'1st_date'},inplace=True)
```

```python
data=data.merge(min_date,on='cust_id')
```

```python
data.head()
```

```python
data['retention']=data.pay_time.dt.month-data['1st_date'].dt.month
# data['actual_date']=data['pay_time'].map(lambda x:x.strftime('%Y-%m'))
data['date']=data['1st_date'].map(lambda x:x.strftime('%Y-%m'))
# data=data[data['gap']<=10]
data.head(10)
```

```python
res=data.pivot_table('cust_id',index='date',columns='retention',aggfunc='count')
res
```

```python
s0=res.iloc[:,0]
scaler=lambda x:x/s0
```

```python
res_1=res.apply(lambda x:x/s0)#apply是对每一列的操作
res_1
```

```python
res_1.fillna(0,inplace=True)
res_p=res_1.applymap(lambda x:format(x,'.2%'))
res_p
```

```python
sns.heatmap(res_1.iloc[:,1:],annot=True,cmap="YlGnBu")
```
